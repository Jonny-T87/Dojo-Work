2.

```python
#changing datetime from object to date series
df['datetime'] = pd.to_datetime(df['datetime'])
```

```python
df['month (name)'] = df['datetime'].dt.month_name()
```

```python
df.head()
```

```python
df['day of week'] = df['datetime'].dt.day_name()
```

```python
df.head()
```

```python
df['hour of day'] = df['datetime'].dt.hour.astype(object)
```

```python
df.head()
```

```python
df.info()
```

```python
#Dropping labels casual and registered
df = df.drop(labels= ['datetime', 'season'], axis=1)
```

3.

```python
def C_to_F(number):
  return (number *9/5) + 32
```

```python
df['temp'] = df['temp'].apply(C_to_F)
```

```python
df['atemp'] = df['atemp'].apply(C_to_F)
```

```python
df.head()
```

4.

```python
df['temp_variance'] = df['atemp'] - df['temp']
```

```python
df.head()
```

```python
#Dropping labels casual and registered
df = df.drop(labels= ['atemp'], axis=1)
```
