```python
filename = '/content/drive/MyDrive/DojoBootCamp/Project Files/RailsToTrails_National Count Data_week.xlsx'
bike_df = pd.read_excel(filename)
```

```python
bike_df.head()
```

```python
#Delete the unnecessary columns 7 by itself
bike_df = bike_df.drop(columns = 'Unnamed: 7')
```

```python
# Delete 2 columns at once 8 & 9
bike_df = bike_df.drop(columns=['Unnamed: 8', 'Unnamed: 9'])
```

```python
bike_df.head()
```

```python
#Rename the Unnamed: 6 column to percent_change.
bike_df = bike_df.rename(columns = {'Unnamed: 6': 'percent_change'})
bike_df.head()
```
