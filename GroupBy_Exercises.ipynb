## Selecting Data

1. Select all data for the date July 24 2005 having flow id=7.
2. From the data selected in step 1, sort the result on Count in descending order and output the top 10 rows. Assign the output to a variable named top10.

```python
July_24 = df['Date'] == '07/24/05'
Flow_7 = df['Flow_ID'] == 7
Flow_7_July_24 = df.loc[Flow_7 & July_24]
```

```python
Flow_7_July_24
```

```python
#https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.sort_values.html used for help
top10 = Flow_7_July_24.sort_values(by=['Count'], ascending=False).head(10)
```

```python
top10
```

## Apply function

```python
top10.describe()
```

```python
#Subtracted the min from .describe in Count 
top10_function  = top10['Count']-1
top10_function
```

```python
#hsow avg of Count from new data
top10_function.mean()
```
